```python
import requests
from bs4 import BeautifulSoup

baseurl = "https://www.adac.de"
url = "https://www.adac.de/rund-ums-fahrzeug/autokatalog/marken-modelle/?filter=NONE&sort=SORTING_DESC"
xpath = "/html/body/div/div[7]/main/div[3]/div/div[1]/a/p"
```

```python
# get ADAC webpage listing brands:
response = requests.get(url).content
soup = BeautifulSoup(response)
```

```python
# trial and error brought us this path to the list of nodes whose parent contains a link to the brand's page:
brand_nodes = soup.html.body.find("main").find_all("p", class_ = "_1LCsObWXtjVxV8x0penxZ-")[3:]
brands = {}
for i, node in enumerate(brand_nodes):
    brandname = node.text
    url = node.parent["href"]
    brands[brandname] = baseurl + url
```
